```python
ports=pd.get_dummies(df3.Embarked, prefix='Embarked')
ports = ports.astype(int)
ports.head(10)
```

```python
df3=df3.join(ports)
```

```python
df3.head()
```

```python
df3=df3.drop(['Embarked'], axis=1)
df3.head()
```

```python
df3.Sex=df3.Sex.map({'male':0, 'female':1})
```

```python
y = df3.Survived.copy()
```

```python
x = df3.drop(['Survived'], axis=1)
```

```python
drop_lst=['Ticket','Name','PassengerId']
```

```python
x.drop(drop_lst,axis=1, inplace=True)
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30) 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier 
```

```python
model_log=LogisticRegression(max_iter=500)
model_log.fit(x_train, y_train)
```

```python
y_pred_log=model_log.predict(x_test)
```

```python
df3.Survived.value_counts()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report
```

```python
accuracy_log = accuracy_score(y_test, y_pred_log)*100
print('Accuracy of logistic model is equal '+str(round(accuracy_log,1))+'%')
```

```python
matrix= confusion_matrix(y_test,y_pred_log, labels=[1,0])
print('Confusion matrix : \n',matrix)
```

```python
matrix_log = classification_report(y_test,y_pred_log,labels=[1,0])
print('Classification report : \n',matrix_log)
```

```python
model_grad = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model_grad.fit(x_train, y_train)
```

```python
y_pred_grad = model_grad.predict(x_test)
```

```python
# Evaluate the model
accuracy_grad = accuracy_score(y_test, y_pred_grad)*100
print('Accuracy of gradient boosting model is equal '+str(round(accuracy_grad,1))+'%')
```

```python
matrix_grad= confusion_matrix(y_test,y_pred_grad, labels=[1,0])
print('Confusion matrix : \n',matrix_grad)
```

```python
matrix_grad = classification_report(y_test,y_pred_grad,labels=[1,0])
print('Classification report : \n',matrix_grad)
```
